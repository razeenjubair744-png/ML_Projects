## Project 1: Sonar Signal Classification

The Sonar dataset contains sonar signal reading collected by bouncing sound waves , with the goal of predicting whether the object is a rock or a mine.

The dataset has 60 numerical columns that represent the strength of sonar signals of sonar signals at different frequencies. The final column is the target label, where R denotes rock and M denotes mine.

## Data Preprocessing

```python
# seperate feature and target
X = df.drop(columns=[60])
y = df[60]
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify = y,
    random_state=random_state
)
```

```python
print("Dataset Shape:", X.shape)
print("Training Dataset Shape:", X_train.shape)
print("Test Dataset Shape:", X_test.shape)
```

```python
# feature scaling
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = SVC()
```

```python
# train the model
model.fit(X_train_scaled, y_train)
```

```python
y_train_pred = model.predict(X_train_scaled)
train_acc = accuracy_score(y_train, y_train_pred)
print(f"Training Data Accuracy: {round(train_acc*100, 2)}%")
```

```python
y_test_pred = model.predict(X_test_scaled)
test_acc = accuracy_score(y_test, y_test_pred)
print(f"Test Data Accuracy: {round(test_acc*100, 2)}%")
```

```python
print("Training Data - Classification Report")
print(classification_report(y_train, y_train_pred))
```

```python
print("Test Data - Classification Report")
print(classification_report(y_test, y_test_pred))
```

```python
def predict_object(input_features):
    # scale features
    scaled_features = scaler.transform([input_features])
    # get prediction from teh model
    prediction = model.predict(scaled_features)
    print("Model prediction", prediction)
    if prediction[0] == 1:
        print("The object is identified as Mine 💣")
    else:
        print("The object is identified as Rock 🪨")
```
